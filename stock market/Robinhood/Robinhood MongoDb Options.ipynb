```python
import pandas as pd
import pymongo
from pymongo import MongoClient
client = MongoClient()
print(client)
db = client['robinhood_data']
print(db)
print(db.list_collections())
print(db.list_collection_names())
#db.drop_collection('ERIC')

import robin_stocks as r 
login = r.login('philip0massey','Rie1Hnd2*#')
```

```python
symbol_list = ['SPY','QQQ','TECL']
expirationDate='2020-02-21'
```

```python
nSymbols = len(symbol_list)
i = 0
while i < nSymbols:
    symbol = symbol_list[i]
    i += 1
    quotes = r.get_quotes(symbol)
    last_trade_price = quotes[0]['last_trade_price']
    updated_at = quotes[0]['updated_at']

    optionData = r.find_options_for_list_of_stocks_by_expiration_date([symbol], expirationDate=expirationDate,optionType='both') 
    l =  list(filter(lambda x:x['delta'] is not None and float(x['delta'])>0.01,optionData))
    df = pd.DataFrame(l)
    df['_id'] = df['expiration_date']+'_'+df['strike_price']+'_'+df['updated_at']
    df['last_trade_price'] = last_trade_price
    df['updated_at'] = updated_at
    data_dict = df.to_dict(orient='records')  # Here's our added param..
    try:
        db[symbol].insert_many(data_dict)
    except pymongo.errors.DuplicateKeyError as e:
        print(e.error_document)
    except Exception as ee:
        print(ee)


    symbol_records = len(list(db[symbol].find({'chain_symbol':symbol})))
    print(symbol,symbol_records,sep='\t')

print("Completed")
```
