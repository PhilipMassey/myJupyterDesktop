```python
import matplotlib.pyplot as plt
from matplotlib import style
import quandl
from matplotlib.pyplot import *
from matplotlib.font_manager import FontProperties
from pandas.plotting import register_matplotlib_converters 
register_matplotlib_converters()
import datetime as dt
import pandas as pd
import pandas_datareader.data as web
 
#print plt.style.available
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [20, 10]
 
fig=plt.figure()
ax1=fig.add_subplot(111)
 
start = dt.datetime.today() - dt.timedelta(days=45)
end = dt.datetime.today()
stock_tickers = ['WMT','HD','COST','M','TJX','KR','JWN','BBY','KSS']
df = web.DataReader(stock_tickers, 'yahoo', start, end)

print("Plotting ...")
 
 
for i in range(len(stock_tickers)):
    try:
        dateaxis = df.reset_index()['Date']
        plt.plot(dateaxis, df.iloc[:,i], label=stock_tickers[i], marker='.')
        #ax1.annotate(stock_tickers[i], xy=(dateaxis[i], series.iloc[:,i][i]), rotation=0)
    except Exception as e:
        print( "error at " + debtarray[i])
        pass
 
plt.xlabel("Year")
plt.ylabel("% change")
 
myFont = FontProperties()
#myFont.set_size('xx-small')
 
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0., numpoints = 1)
legend = ax1.legend(bbox_to_anchor=(0, 0, 1, 1), loc=0, ncol=1,
           prop = myFont,fancybox=True,shadow=False,title='Debt as % of GDP')
plt.setp(legend.get_title(),fontsize='small')
plt.show()
```
