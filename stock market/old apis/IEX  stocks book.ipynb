```python
import matplotlib.pyplot as plt
import matplotlib.dates as dates
%pylab inline
pylab.rcParams['figure.figsize'] = (12, 8)   # Change the size of plots

import pandas as pd
import pandas_datareader as web
import datetime

tickers = ['VCR','VDC','VFH','VGT','VHT','VIS','VNQ','VPU']

dbroker = { 'VCR':{'count':65, 'cost': 164.26,  'stop': 164.26},
            'VDC':{'count':85, 'cost': 130.34,  'stop': 70.26},
            'VFH':{'count':160, 'cost': 71.97,  'stop': 70.26},
            'VGT':{'count':65, 'cost': 182.02, 'stop':173.31},
            'VHT':{'count':70, 'cost': 158.56, 'stop':157.30},
            'VIS':{'count':145, 'cost': 145.52,  'stop': 76.67},
            'VNQ':{'count':145, 'cost': 76.83,  'stop': 76.67},
            'VPU':{'count':95, 'cost': 110.56, 'stop':115.43},
            'VBR':{'count':80, 'cost': 136.58, 'stop':115.43}
          }
tickers = list(dbroker.keys())
total_cost = 0
for tick in tickers:
    stock = dbroker[tick]
    stock_tcost = stock['count'] * stock['cost']
    total_cost += stock_tcost
print('total cost:\t{0:.2f}'.format(total_cost))
```

```python
import csv
dbroker
w = csv.writer(open("dbroker.csv", "w"))
for key, val in dbroker.items():
    w.writerow([key, val])
import pickle
def save_obj(obj, name ):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
    
#save_obj(dbroker, 'dbroker')
dbroker = load_obj('dbroker')
dbroker
```

```python
tickers = list(dbroker.keys())

stocks = web.DataReader(tickers, 'iex-last')

total_change = 0.0
print('{0}\t\t{1}\t\t{2}\t\t{3}\t\t{4}'.format( 'ticker','price','uchange','tchange','til_stop'))

for tick in tickers:
    price = stocks[stocks['symbol'] == tick]['price'].values[0]
    cost = dbroker[tick]['cost']
    if price == 0:
        price = cost
    uchange = price - cost
    count = dbroker[tick]['count']
    tchange = count * uchange
    total_change += tchange
    stop = dbroker[tick]['stop']
    til_stop = price - stop

    print('{0}\t\t{1:.2f}\t\t{2:.2f}\t\t{3:.2f}\t\t{4:.2f}'.format(tick,price,uchange, tchange, til_stop))

print('\ntotal change \t\t\t\t\t{0:.2f}'.format(total_change))
```
