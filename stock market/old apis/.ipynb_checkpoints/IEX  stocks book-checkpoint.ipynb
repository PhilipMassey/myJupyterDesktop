```python
import matplotlib.pyplot as plt
import matplotlib.dates as dates
%pylab inline
pylab.rcParams['figure.figsize'] = (12, 8)   # Change the size of plots

import pandas as pd
import pandas_datareader as web
import datetime
 
```

```python
dbroker = { 'VFH':{'count':1000, 'cost': 70.26,  'stop': 70.26},
            'VGT':{'count':1000, 'cost': 173.31, 'stop':173.31},
            'VHT':{'count':1000, 'cost': 157.30, 'stop':157.30},
            'VPU':{'count':1000, 'cost': 115.43, 'stop':115.43},
            'VTV':{'count':1000, 'cost':104.43,  'stop':104.43},
            'VYM':{'count':1000, 'cost':83.09,   'stop': 83.09},
            'VEA':{'count': 1000, 'cost':45.03,  'stop': 45.03},
            'VWO':{'count':2000,  'cost':45.46,  'stop': 45.46},
           'VWOB':{'count':1000,  'cost':77.07,  'stop': 77.07}
          }
ticker = list(dbroker.keys())

stocks = web.DataReader(ticker, 'iex-last')

total_change = 0.0
print('{0}\t\t{1}\t\t{2}\t\t{3}\t\t{4}'.format( 'ticker','price','uchange','tchange','til_stop'))

for tick in ticker:
    price = stocks[stocks['symbol'] == tick]['price'].values[0]
    cost = dbroker[tick]['cost']
    if price == 0:
        price = cost
    uchange = price - cost
    count = dbroker[tick]['count']
    tchange = count * uchange
    total_change += tchange
    stop = dbroker[tick]['stop']
    til_stop = price - stop

    print('{0}\t\t{1:.2f}\t\t{2:.2f}\t\t{3:.2f}\t\t{4:.2f}'.format(tick,price,uchange, tchange, til_stop))

print('\ntotal change \t\t\t\t\t{0:.2f}'.format(total_change))

start = datetime.datetime(2018,4,1)
end = datetime.date.today()


```
