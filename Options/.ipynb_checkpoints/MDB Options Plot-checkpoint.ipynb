```python
import plotly.graph_objs as go
import plotly.figure_factory.utils as utils
import pandas as pd
from pymongo import MongoClient
client = MongoClient()
db = client['robinhood_options']
```

```python
ticker = 'SDGR'
df = pd.DataFrame(list(db[ticker].find({'chain_symbol': ticker})))
print(df.columns)
```

```python
df.index
```

```python
df[['strike_price', 'last_trade_price','ticker_last_trade_price']] = df[['strike_price','last_trade_price' ,'ticker_last_trade_price']].apply(pd.to_numeric)
df[['ticker_updated_at','updated_at']] = df[['ticker_updated_at','updated_at']].apply(pd.to_datetime)
```

```python
print(set(df['strike_price']))
```

```python
#x = list(filter(lambda l: l['strike_price']==22.5,df))
fdf = df[(df.strike_price == 55.00) & (df.type == 'call')]
```

```python

fig = go.Figure()
fig.add_trace(go.Scatter(x=fdf.ticker_updated_at, y=fdf['ticker_last_trade_price'], name="stock price",
                         line_color='dimgray'))

fig.update_layout(title_text='Ticker Price',
                  xaxis_rangeslider_visible=True)
fig.show()
```

```python

fig = go.Figure()
fig.add_trace(go.Scatter(x=fdf.ticker_updated_at, y=fdf['last_trade_price'], name='option price', line_color='deepskyblue'))
# fig.add_trace(go.Scatter(x=fdf.ticker_updated_at, y=df['ticker_last_trade_price'], name="stock price",line_color='dimgray'))
fig.update_layout(title_text='Strike Price $55.00', xaxis_rangeslider_visible=True)
fig.show()
```
