```python
import pandas as pd
from datetime import datetime
import os
import robin_stocks as r 
login = r.login(os.getenv('RHUSER'), os.getenv('RNPWD'))
import plotly.graph_objects as go
```

```python
symbol ='SPY'
print(r.get_name_by_symbol(symbol))
list_of_dicts = r.get_historicals(symbol,span='year')
df = pd.DataFrame.from_dict(list_of_dicts, orient='columns')
from datetime import datetime
df['Date']= (pd.to_datetime(df['begins_at'], format="%Y-%m-%dT%H:%M:%SZ"))
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['open_price'],
                high=df['high_price'],
                low=df['low_price'],
                close=df['close_price'])])
fig.show()
```
